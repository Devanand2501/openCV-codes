```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

* We can vice-versa operation similarly

```python
img = cv.imread('Resources/Photos/cats 2.jpg')
cv.imshow('Colored Image', img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## BGR TO GRAY

```python
gray_img = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow('Colored Image', img)
cv.imshow('Gray Image', gray_img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## BGR to HSV

```python
hsv_img = cv.cvtColor(img,cv.COLOR_BGR2HSV)
cv.imshow('Colored Image', img)
cv.imshow('Hsv Image', hsv_img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## BGR TO LAB

```python
lab_img = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow('Colored Image', img)
cv.imshow('Lab image',lab_img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## BGR TO RGB

OpenCV display images in BGR format 
Matplotlib display images in RGB format


```python
plt.imshow(img)
plt.show()
```

```python
rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
cv.imshow('Colored Image', img)
cv.imshow('rgb Image', rgb)
cv.waitKey(5000)
cv.destroyAllWindows()

plt.imshow(rgb)
plt.show()
```
