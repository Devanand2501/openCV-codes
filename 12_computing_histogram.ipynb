* Histogram is calculated to shows how many pixels in the image have a specific intensity value.

```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Resources/Photos/cats.jpg')
cv.imshow('Colored image',img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## Grayscale Histogram

```python
cv.calcHist?
```

2nd parameter is channel where [0] for blue, [1] for green, and [2] for red.

```python
gray = cv.cvtColor(img,cv.COLOR_BGRA2GRAY)
cv.imshow('Gray image',gray)
cv.waitKey(5000)
cv.destroyAllWindows()
```

```python
gray_hist = cv.calcHist([gray],[0],None,[256],[0,256])
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels ")
plt.xlim([0,256])
plt.plot(gray_hist)
plt.show()
```

## Colored histogram

```python
blank = np.zeros(img.shape[:2],dtype='uint8')
circle = cv.circle(blank,(img.shape[1]//2,img.shape[0]//2),100,255,-1)

mask = cv.bitwise_and(img,img,mask=circle)
cv.imshow('Masked image',mask)
cv.waitKey(5000)
cv.destroyAllWindows()
```

```python
colors =('b','g','r')
for i,col in enumerate(colors):
    hist = cv.calcHist([img],[i],circle,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim([0,256])
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels ")
plt.show()
```
