```python
# For reading videos 
# We can pass the integer argument if we have cameras connected (mostly 0 if multiple then accordingly)
# If we can also pass the file path also
capture = cv.VideoCapture ('Resources/Videos/dog.mp4')
```

#### For reading videos

```python
while True:
    isTrue, frame = capture.read()
    cv.imshow('Video',frame)
    if cv.waitKey(20) & 0xFF == ord('d'): #If the letter d is pressed it will break out of while loop
        break

# Release the video capture object
capture.release()
cv.destroyAllWindows()
#Error is occured when video gets over
```
