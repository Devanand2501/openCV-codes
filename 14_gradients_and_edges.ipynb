```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

* Canny edge detector is more cleaned version of Sobel

```python
img = cv.imread('Resources/Photos/park.jpg')
cv.imshow("Colored image",img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

```python
gray = cv.cvtColor(img,cv.COLOR_BGRA2GRAY)
cv.imshow("Gray image",gray)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## Laplacian

```python
lap = cv.Laplacian?
```

```python
lap = cv.Laplacian(gray,cv.CV_64F)
lap = np.uint8(np.absolute(lap))
cv.imshow("Colored image",img)
cv.imshow("Laplacian image",lap)
plt.imshow(lap)
plt.axis('off')
cv.waitKey(5000)
cv.destroyAllWindows()
```

## Sobel

Gradient is calculated in x as well as y direction

* Sobel X

```python
sobel_x = cv.Sobel(gray,cv.CV_64F,1,0)
cv.imshow("Sobel X image",sobel_x)
plt.imshow(sobel_x)
plt.title("Sobel X")
plt.axis('off')
cv.waitKey(5000)
cv.destroyAllWindows()
```

* Sobel Y

```python
sobel_y = cv.Sobel(gray,cv.CV_64F,0,1)
cv.imshow("Sobel Y image",sobel_y)
plt.imshow(sobel_y)
plt.title("Sobel Y")
plt.axis('off')
cv.waitKey(5000)
cv.destroyAllWindows()
```

```python
combined_sobel = cv.bitwise_or(sobel_x,sobel_y)
# combined_sobel = cv.Sobel(gray,cv.CV_64F,1,1)
cv.imshow("combined sobel image",combined_sobel)
cv.waitKey(5000)
cv.destroyAllWindows()
```
